# Values
We'll get more details on this later, but a value is just an individual piece of data. For now, we'll focus on a few "kinds" or "types" of data:
1. Integer Numbers (int)
2. Decimal Numbers (float)
3. Strings (str)

- You can tell when something is an `int` because it's a number without a decimal place.
- You can tell when something is a `float` because it's a number with a decimal place.
- You can tell when something is a `str` because it's inside of either double- or single-quotation marks:
`'Hello'` or
`"Hello"`

# Expressions 
An expression is a chunk of code that will "evaluate" to a value (i.e. turn into a single value).That is to say, anything that results in a single value after some computation is anexpression. Here are some examples:


```python
42             # This is a value

"Hello, world" # This is a value

42*2           # This is an expression that turns into the value 84

"Goodbye, "+ "world".    # This is an expression that turns into the value "Goodbye, world"
```

The most basic way of creating expressions is to combine values using operators.

# Colors in Python
* We can specify colors as 3 numbers
  * One for R
  * One for G
  * One for B

```python
import dudraw
#Create our canvas
dudraw.set_canvas_size(400,400)
dudraw.set_x_scale(0,400)
dudraw.set_y_scale(0,400)
#Set the Pen Color
dudraw.set_pen_color_rgb(0,255,0)
#Draw the circle
dudraw.filled_circle(200, 200, 100)
dudraw.set_pen_color_rgb(255,0,0)
dudraw.filled_circle(200, 200, 50)
dudraw.show(100000)
```
